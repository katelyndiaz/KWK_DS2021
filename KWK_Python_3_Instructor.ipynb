## ü•ß **NumPy**

**What is NumPy?**

NumPy is a Python library used for working with arrays.It also has functions for working in domain of linear algebra, fourier transform, and matrices. NumPy was created in 2005 by Travis Oliphant. It is an open source project and you can use it freely. NumPy stands for Numerical Python.

<br>

**Why Use NumPy?**

In Python we have lists that serve the purpose of arrays, but they are slow to process. NumPy aims to provide an array object that is up to 50x faster than traditional Python lists. The array object in NumPy is called ndarray, it provides a lot of supporting functions that make working with ndarray very easy. Arrays are very frequently used in data science, where speed and resources are very important.

<br>

**Why is NumPy Faster Than Lists?**

NumPy arrays are stored at one continuous place in memory unlike lists, so processes can access and manipulate them very efficiently. This behavior is called locality of reference in computer science. This is the main reason why NumPy is faster than lists. Also it is optimized to work with latest CPU architectures.

[(source and more info)](https://www.w3schools.com/python/numpy_intro.asp)

### **NumPy Arrays**

An array is a central data structure of the NumPy library. It‚Äôs a grid of values and it contains information about the raw data, how to locate an element, and how to interpret an element. It has a grid of elements that can be indexed in various ways. The elements are all of the same type, referred to as the array dtype (data type).

The number of dimensions and items in an array is defined by its shape. The shape of an array is a tuple of non-negative integers that specify the sizes of each dimension. In NumPy, dimensions are called axes. This means that if you have a 2D array that looks like this:

```
[[0., 0., 0.],
 [1., 1., 1.]]
 ```
The array above has 2 axes. The first axis has a length of 2 and the second axis has a length of 3.

### **Creating a Pandas DataFrame**

The two important structures for pandas is the series and the dataFrame. 

*   **Series:**  A series is a one-dimensional object that can hold any data type such as integers, floats and strings. Let‚Äôs take a list of items as an input argument and create a Series object for that list. You can think of this as just one column.
*   **DataFrame:** A dataframe is a two dimensional object that can have columns with potential different types. Different kind of inputs include dictionaries, lists, series, and even another DataFrame. You can think of this as a whole google sheet!

Now we're going to create our first dataframe! Let's start by making some data:

````
# Creating data using lists 
names = ['Karlie', 'Taylor', 'Billie', 'Lizzo', 'Demi']
age = [ 28, 31, 19, 32, 28]
city = ["Chicago", "Wyomissing", "Los Angeles", "Detroit", "Dallas"]
````

````
# deleting a column 
new_friends.drop('age', axis=1) 

# Note: axis=1 denotes that we are referring to a column, not a row
````

Notice the code cell above returned our dataframe without the "age" column. However, if we print out our "new_friends" dataframe again in a new cell, this change will not be saved.

````
print(new_friends)
````

In order to save the changes made, we have to tell Pandas that we want to replace our old dataframe. We can do this by setting it equal to itself ex. `new_friends = new_friends`

````
# deleting the city column 
new_friends = new_friends.drop('age', axis=1)
print(new_friends)
````

````
# deleting a row using the index number 
new_friends.drop(1, axis=0)

# Note: axis=0 denotes that we are referring to a row, not a column
````

We can specify which columns we want to see using this syntax: `dataframe_name['column_name']`

````
new_friends['names']
````

````
# show multiple columns in a dataframe
new_friends[['names', 'city']]
````

### **Filtering With Pandas**

Pandas lets you subset a dataframe based on one or more values of a specific column. Essentially, we would like to select rows based on one value or multiple values present in a column.

Showing only the rows where the age is greater than 30:

# üßê **Data Cleaning & Exploration**

## üßΩ **Data Cleaning & Preparation**

Data cleaning sounds like an boring task, but in reality, it is one of the most important things you will do in data science. Common points to consider when data cleaning include:

- **Formatting** - is our data presented in a way that makes sense? Are the variable types correct? (e.g.: is there a column of strings that are supposed to represent numbers?)
- **Missing data** - are there any rows or columns that have blank, np.NaN, or otherwise missing data? Should they be cut out? 

Cleaning the data predominantly takes place at the beginning of an analysis, but often times you have to inspect your data to learn what needs to be cleaned.

### **Formatting** 

There are many methods for processing unstructured data and various functions that can help us acheive this include: 

**I. Selecting and Renaming Columns and Rows:**

- **`rename()`** =  Method that accepts a dictionary mapper or function to apply to the column or index names. 
- **`columns`** = Dataframe attribute returning the column labels. Can be directly overwritten.
- **`filter()`** = Dataframe Method to select specific columns or indexes from a list like argument or based on a regex expression which matches the column or row labels.
- **`drop()`** = Dataframe Method to drop specified columns or rows by label. Can be set to perform inplace.

**II. Working With Data Types:**
- **`pd.to_numeric()`** = Pandas function to convert argument to a numeric type. Can be applied to Series, list, 1d array, or scalar.
- **`astype()`** = Method to convert a dataframe or pandas series to a specified data type.
- **`infer_objects()`** - Dataframe Method that attempts to infer better data types for object columns.

**III. Removing Duplicates, Mapping Functions, and Replacing Values:**
- **`duplicated()`** = Dataframe, Series, and Index method to identify duplicate rows via a boolean series.
- **`drop_duplicates()`** = Dataframe Method that removes the duplicated rows. Can be set to perform inplace. 
- **`np.unique()`** = A numpy function that can find the unique elements of an array. 
- **`map()`** = Series method to perform element wise operators or substitutions across the series. 
- **`replace()`** = Dataframe, Series, and string Method to replace a value with another value 
- **`where()`** = Dataframe and Series method to replace values based on a boolean condition. Similar to the function `np.where()`

**Working with Strings**
- **`str`** = Pandas Series method that allows for use of built in python string methods to be vectorized. NA's will stay NA's.
- **`str[val]`** = Series method to slice each string element in the series |
- **`str.lower()`** = An example of using the built in python lowercase method on the entire series|





### **Missing Data** 

One of the most common problems in data cleaning is missing data. In nearly all statistical methods or machine learning models, it is assumed that the examples have information on every variable included. This is often not the case!

General rules of thumb when imploying these strategies are as follows:
- If the missing data is less than 3% of the total obseravations it doesn't make much of a difference in your approach
- If the proportion of missing data is greater than 3% and your data is not missing completely at random (e.g., clerical error, illegible writing in a survey response, dropped the test tube in the lab.), listwise and pairwise deletion will result in biased results and imputation is preferred
- Depending on the data, 50% of missing data can be a good threshold to remove the variable entirely, sometimes it is 25%. What matters more is how important that variable is to the analysis. 
- If the goal of your analysis is prediction, try various strategies and compare results

## üßê **Exploring Data**

In stats, exploratory data analysis (EDA) is an approach to analyzing data sets to summarize their main characteristics. It's a way of asking questions of your data to better understand it. 

The process of an EDA can be thought of as preparing a rough draft of your data analysis. We will cover techniques and methods for this process but throughout these methods keep in mind the goals of an EDA listed below:
- Check for problems/inconsistencies with the collected data, such as missing data or errors in the measurements
- Understand the variable types, summary statistics, and distributions
- Identify relationships between variables that are interesting or unexpected
- Confirm the validity of your question by looking for evidence that matches your expectations
- Determine if your question can be answered with the data or if you need to collect more data
